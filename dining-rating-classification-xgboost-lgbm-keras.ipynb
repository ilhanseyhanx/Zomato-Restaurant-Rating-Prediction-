```python
import numpy as np 
import pandas as pd 

import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
dataset = pd.read_csv("/kaggle/input/zomato-restaurant-dataset/enhanced_zomato_dataset_clean.csv")
dataset.head(5)
```

```python
dataset.info()
```

```python
dataset.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder

label_cols = ["Restaurant_Name", "Cuisine", "City", "Item_Name", "Best_Seller","Place_Name"]

for col in label_cols:
    dataset[col] = LabelEncoder().fit_transform(dataset[col].astype(str))
```

```python
dataset.head()
```

```python
from sklearn.model_selection import train_test_split

x = dataset.drop(columns = ["Dining_Rating"])
y = dataset["Dining_Rating"]

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.2,random_state = 0)
```

```python
from xgboost import XGBRegressor as XGB
from sklearn.metrics import mean_squared_error,r2_score,mean_absolute_error

model_xgb = XGB()
model_xgb.fit(x_train,y_train)
y_pred_xgb = model_xgb.predict(x_test)
print("MSE: ",mean_squared_error(y_test,y_pred_xgb))
print("R2 SCORE: ",r2_score(y_test,y_pred_xgb))
print("MAE: ",mean_absolute_error(y_test,y_pred_xgb))
```

```python
from lightgbm import LGBMRegressor as LGBM

model_lgbm = LGBM()
model_lgbm.fit(x_train,y_train)
y_pred_lgbm = model_lgbm.predict(x_test)
print("MSE: ",mean_squared_error(y_test,y_pred_lgbm))
print("R2 SCORE: ",r2_score(y_test,y_pred_lgbm))
print("MAE: ",mean_absolute_error(y_test,y_pred_lgbm))
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.optimizers import Adam
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
x_train_scaler = scaler.fit_transform(x_train)
x_test_scaler = scaler.transform(x_test)

model_keras = Sequential([
    Dense(128,activation = "relu",input_shape=(x_train_scaler.shape[1],)),
    Dropout(0.3),
    Dense(64,activation="relu"),
    Dropout(0.3),
    Dense(1)
])

model_keras.compile(optimizer = Adam(),loss = "mse",metrics = ["mae"])
history = model_keras.fit(
    x_train_scaler,y_train,
    validation_data = (x_test_scaler, y_test),
    epochs = 20,
    batch_size = 128,
    verbose = 1
)
```

```python
y_pred_keras = model_keras.predict(x_test_scaler).flatten()
print("R2 SCORE: ",r2_score(y_test,y_pred_keras))
print("MSE: ",mean_squared_error(y_test,y_pred_keras))
print("MAE: ",mean_absolute_error(y_test,y_pred_keras))
```
